```
model = linearRegression(x_data.shape[1])

lr = 0.005
epochs = 30

loss_func = torch.nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr)
```

```
trainLossList = []
```

```
#Shift everything to tensors
inputs = torch.from_numpy(x_data)
targets = torch.from_numpy(y_data)
print(inputs.dtype, targets.dtype)
```

```
#Models want things as floats, not doubles
inputs = torch.tensor(inputs, dtype=torch.float32) 
targets = torch.tensor(targets, dtype=torch.float32).reshape(-1, 1)
```

```
for i in range(epochs):
    model.train()
    optimizer.zero_grad()
    outputs = model(inputs)

    loss = loss_func(outputs, targets)
    trainLossList.append(loss.item())
    loss.backward()
    optimizer.step()
    print(outputs[0], targets[0])
    print("Epoch:", i, "Loss:", loss.item())
```

```

plt.figure(figsize=(12, 6))

plt.plot(trainLossList)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.show()
```

```
from sklearn.metrics import accuracy_score

model.eval()
correct = 0
with torch.no_grad():
    outputs = model(inputs)
    predicted = torch.round(outputs).squeeze().detach().numpy() 
    predicted = predicted.astype(int)  
    accuracy = accuracy_score(targets.numpy(), predicted)
    print(f'Accuracy on Test Set: {accuracy * 100:.2f}%')
```
